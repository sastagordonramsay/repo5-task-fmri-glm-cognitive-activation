```python
import os
import pandas as pd
```

```python
os.makedirs("data/raw", exist_ok=True)
os.makedirs("data/processed", exist_ok=True)
os.makedirs("figures", exist_ok=True)

print("Folders ready")
```

```python
for root, dirs, files in os.walk("data/raw"):
    for file in files:
        print(os.path.join(root, file))
```

```python
participants = pd.read_csv(
    "data/raw/participants.tsv",
    sep="\t"
)

participants.head()
```

```python
print("Total participants:", len(participants))
print("\nColumns:")
print(participants.columns.tolist())
```

```python
candidates = pd.DataFrame({
    "participant_id": participants["participant_id"].tolist()[:80]
})

candidates.head()
```

```python
candidates.to_csv(
    "data/processed/candidate_subjects.csv",
    index=False
)

print("Saved candidate_subjects.csv")
```
